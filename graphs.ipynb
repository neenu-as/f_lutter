```python
from matplotlib import pyplot as plt
x=[10,12,15,17]

plt.hist(x,ec="red",bottom=10,rwidth=3)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("histogram")
plt.show
```

```python

import matplotlib.pyplot as plt
x=[15,25,12,6]
y=["excellent","good","average","poor"]
plt.figure(figsize=(8,10))
plt.pie(x,labels=y,startangle=0,explode=[0.2,0,0,0],colors=["red","yellow","blue","green"],autopct="%2.1f%%")
plt.legend(title="performance")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["science","commerce","arts"]
h=[200,100,150]
w=[0.2,0.8,0.5]
c=["red","green","blue"]
b=[25,33,12]
plt.bar(x,h,width=w,color=c,bottom=b)
plt.xlabel("courses")
plt.ylabel("student enrolled")
plt.title("bar chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
w=0.2
x=["usa","china","india","germany"]
gold=[100,120,300,60]
silver=[20,28,49,40]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
plt.bar(bar1,gold,w,label="gold")
plt.bar(bar2,silver,w,label="silver")
plt.xlabel("countries")
plt.ylabel("medels")
plt.title("medel vs countries")
plt.xticks(bar1+w)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
w=0.2
x=[10,20,30,40,50,]
men=[22,30,35,35,26]
women=[25,32,30,35,29]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
plt.bar(bar1,men,w,label="men")
plt.bar(bar2,women,w,label="women")
plt.xlabel("gender")
plt.ylabel("scores")
plt.title("scores buy group and gender")
plt.xticks(bar1+w)
plt.legend()
plt.show()
```
